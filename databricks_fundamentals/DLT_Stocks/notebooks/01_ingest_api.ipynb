# 01 - Ingest Stock Data

This notebook calls the Alpha Vantage API and stores raw JSON data in a Unity Catalog Volume.

This represents the raw data ingestion step (Bronze layer input).

```python
import requests
import json
from datetime import datetime

API_KEY = "Your_Alpha_Vantage_API_Key"

symbols = ["IBM", "AAPL", "MSFT"]

for symbol in symbols:
    url = (
        "https://www.alphavantage.co/query"
        f"?function=TIME_SERIES_DAILY"
        f"&symbol={symbol}"
        f"&apikey={API_KEY}"
    )

    response = requests.get(url)
    data = response.json()

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    path = f"/Volumes/workspace/stock_dlt/raw_data/{symbol}_{timestamp}.json"

    dbutils.fs.put(path, json.dumps(data), overwrite=True)

    print(f"Saved {symbol}")
```
